```python
df.drop_duplicates(
    inplace=True
)
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
)
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
)
```

```python
df.replace(
    ['', ' ', 'NA', 'N/A', 'null', 'NULL'],
    np.nan,
    inplace=True
)
```

```python
numeric_columns = [
    'total_sessions',
    'total_orders',
    'total_spend',
    'average_order_value',
    'cart_abandonment_rate',
    'average_discount_used',
    'customer_satisfaction',
    'preferred_device',
    'city',
    'preferred_category',
    'total_products_purchased',
    'total_pages_viewed',
    'average_time_on_site',
    'total_review_votes',
    'preferred_payment_method',
    'rn',
    'churn'
]

for column in numeric_columns:
    df[column] = pd.to_numeric(
        df[column],
        errors='coerce'
    )
```

```python
df.isnull().sum()
```

```python
for column in numeric_columns:
    df[column].fillna(
        df[column].median(),
        inplace=True
    )
```
